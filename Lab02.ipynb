## Modelamiento

```python
# Instalar libreria PyCaret
!pip install pandas==2.1.4 matplotlib==3.7.5 # Por errores
!pip install pycaret
```

```python
# Cargamos data con esta librería
from pycaret.datasets import get_data
data = get_data('diabetes')
```

```python
from pycaret.classification import ClassificationExperiment
s = ClassificationExperiment()
s.setup(data, target = 'Outcome', session_id = 123)
```

```python
best = s.compare_models()
```

## Evaluación

```python
s.evaluate_model(best)
```

```python
pred_holdout = s.predict_model(best)
```

```python
new_data = data.copy().drop('Outcome', axis = 1)
predictions = s.predict_model(best, data = new_data)
```

```python
s.save_model(best, 'best_pipeline')
```

```python
predictions
```

### Podemos visualizar que el modelo ya esta funcionando, según los resultados tiene un 0.76% de precisión, a simple vista vemos que se parece a los resultado de la data original.
